```python
#Menambahkan brightness dan darkness
%matplotlib inline

img = cv2.imread('catering by jimila.png')
M = np.ones(img.shape, dtype='uint8') * 75
img_bright = cv2.add(img, M)
img_dark = cv2.subtract(img, M)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))


ax1.set_title('Brightness')
ax1.imshow(img_bright)
ax2.set_title('Darkness')
ax2.imshow(img_dark)
```

```python
img = cv2.imread('catering by jimila.png')
plt.imshow(img),plt.title('Original')
flip = cv2.flip(img, 1)
flip2 = cv2.flip(img, 0)
f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))


plt.imshow(img)
ax1.set_title('rotasi samping')
ax1.imshow(flip)
ax2.set_title('rotasi bawah')
ax2.imshow(flip2)
```
